```python
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
data = sns.load_dataset("penguins")
```

```python
plt.rcParams.update({
    "figure.dpi": 600,
    "savefig.dpi": 600,
    "figure.figsize": (5, 5),
    "axes.titlesize": 10,
    "axes.labelsize": 8,
    plt.legen
    
})
```

```python
print(data.columns)
```

```python
sns.set_theme(style='whitegrid', palette='tab10')
fig, ax = plt.subplots(figsize=(5, 5), dpi=600)
sns.scatterplot(data, x="bill_length_mm", y="bill_depth_mm", 
                hue="species", ec='black', linewidth=0.5,
                s=80,  # Adjust point size
                alpha=1.0,  # Ensure full opacity
                ax=ax)
ax.grid(True, alpha=0.3, linewidth=0.5)
plt.show()
```

```python
colors = {'Adelie': '#E41A1C', 'Chinstrap': '#377EB8', 'Gentoo': '#4DAF4A'}
sns.scatterplot(data, x="bill_length_mm", y="bill_depth_mm", 
                hue="species", palette=colors, ec='black', 
                linewidth=0.5, s=80, alpha=1.0)
sns.despine(left = True, bottom = True)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left', 
           frameon=False, fancybox=False, shadow=False, facecolor='white')
plt.show()
```

```python
sns.set_theme(style='whitegrid', palette='tab10')
fig, ax = plt.subplots(figsize=(5, 5), dpi=600)
sns.scatterplot(data, x="bill_length_mm", y="bill_depth_mm", 
                hue="species", ec='black', s = 80)
ax.set_frame_on(0)
plt.legend(bbox_to_anchor=(1.05, 0.5), frameon=0, title = "Species")
plt.show()
```
